```python
import pandas as pd
import sklearn.linear_model
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier

from sklearn.cluster import estimate_bandwidth
from sklearn.cluster import MeanShift
import numpy as np
```

# Part 1

```python
# Get data
data = pd.read_csv("https://raw.githubusercontent.com/Micniks/Python-Assignments/master/Week11/cultists_v2.csv")
data
```

```python
# Get the skills
xs = data[['Stealth', 'Influence', 'Endurance', 'Lore', 'Economic', 'Strength', 'Insanity']]
xs
```

```python
# Map roles to a number - as to chew it.
ys = data['Position'].map({'Priest':0,'Enforcer':1,'Assassin':2,'Recruiter':3,'Accountant':4,'Advisor':5})
ys
```

```python
model = sklearn.linear_model.LinearRegression()
model
```

```python
size = 0.25
x_train,x_test,y_train,y_test = train_test_split(xs,ys,test_size=size)
```

```python
model.fit(x_train,y_train)
print("Train XY:", model.score(x_train,y_train))
print("Test XY:", model.score(x_test,y_test))
```

# Part 3

```python
prunned_data = data.drop(['Stealth','Influence','Endurance','Lore','Economic','Strength', 'Insanity','Cultist','Position'], axis='columns')
prunned_data
```

```python
prunned_data = pd.get_dummies(prunned_data,columns=['Type_of_living_area'])
prunned_data
```

```python
prunned_data.dropna(axis='index',inplace=True)
prunned_data
```

```python
estimated_bandwidth = estimate_bandwidth(prunned_data)
estimated_bandwidth
```

```python
analyzer = MeanShift(bandwidth=300) 
analyzer.fit(prunned_data)
labels = analyzer.labels_
print('Clusters:',len(np.unique(labels)))
```

```python
prunned_data['cluster_grp'] = np.nan
for i in range(len(prunned_data)):
    prunned_data.iloc[i,prunned_data.columns.get_loc('cluster_grp')] = labels[i]
prunned_cluster_data = prunned_data.groupby(['cluster_grp']).size()
prunned_cluster_data['Counts'] = pd.Series(prunned_data.groupby(['cluster_grp']).size())
prunned_cluster_data
```
