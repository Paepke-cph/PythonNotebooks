```python
import random
import urllib.request
import gzip
import shutil
import networkx as nx
```

```python
url = "https://snap.stanford.edu/data/wiki-Vote.txt.gz"
urllib.request.urlretrieve(url, "./downloads/Wiki-Vote.txt.gz")

with gzip.open("./downloads/Wiki-Vote.txt.gz", "rb") as f_in:
    with open("./downloads/Wiki-Vote.txt", "wb") as f_out:
        shutil.copyfileobj(f_in, f_out)

graph = nx.read_edgelist("./downloads/Wiki-Vote.txt")
print(nx.info(graph))
"50" in graph
```

```python
def randomGid(graph):
    return random.choice(list(graph.nodes()))

randomGraph = randomGid(graph)
print(randomGraph)
graph[randomGraph]
```

```python
H = graph.subgraph(graph[randomGraph])
nx.draw_circular(H, with_labels=True)
```
